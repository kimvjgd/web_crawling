```python
import requests
from bs4 import BeautifulSoup
# import pandas as pd

response = requests.get("https://www.nature.com/search?q=porous+pdms")
html = response.text
soup = BeautifulSoup(html, "html.parser")
# print(soup)
articles = soup.select(".app-article-list-row__item")
```

```python
for article in articles:
    title = article.select_one(".c-card__title > a").text.strip()
    link = "https://www.nature.com" + article.select_one(".c-card__title > a").attrs['href']
    creator = article.select_one(".c-author-list").text
    date_tag = article.select_one('time[datetime]')
    date = date_tag['datetime'] if date_tag else None
    journal_title_tag = article.select_one('div[data-test="journal-title-and-link"]')
    journal_title = journal_title_tag.get_text(strip=True) if journal_title_tag else None
    volume_info_tag = article.select_one('div[data-test="volume-and-page-info"]')
    volume_info = volume_info_tag.get_text(strip=True) if volume_info_tag else None

    print(title, creator, date, journal_title, volume_info)
```
